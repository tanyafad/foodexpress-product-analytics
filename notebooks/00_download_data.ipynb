```python
from datasets import load_dataset
import pandas as pd
from pathlib import Path
```

```python
customers = load_dataset("mindweave/food-delivery-orders", "customers")
drivers = load_dataset("mindweave/food-delivery-orders", "drivers")
orders = load_dataset("mindweave/food-delivery-orders", "orders")
restaurants = load_dataset("mindweave/food-delivery-orders", "restaurants")
```

```python
customers_df = customers["train"].to_pandas()
customers_df.to_csv("../data/raw/customers.csv", index=False)

drivers_df = drivers["train"].to_pandas()
drivers_df.to_csv("../data/raw/drivers.csv", index=False)

orders_df = orders["train"].to_pandas()
orders_df.to_csv("../data/raw/orders.csv", index=False)

restaurants_df = restaurants["train"].to_pandas()
restaurants_df.to_csv("../data/raw/restaurants.csv", index=False)
```
